### Performing EDA:

```python
df.columns=df.columns.str.strip() ## removing extra spaces from column names
```

```python
df.columns
```

```python
### We will calculate Estimated Unemployment Rate (%) by Region


a=pd.DataFrame(df.groupby('Region')['Estimated Unemployment Rate (%)'].agg('sum').reset_index())
```

```python
a
```

```python
## Now we will plot graph b/w Estimated Unemployment Rate (%) and Region

plt.figure(figsize=(10,6))
plt.plot(a['Region'],a['Estimated Unemployment Rate (%)'],marker='o')
plt.xlabel('Regions')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate vs Region')
plt.xticks(rotation=90)
plt.show()
```

### Sikkim has Less Unemployment Rate and Tripura has high

```python
len(a['Region'])
```

```python
z=[0.1 for _ in range(len(a['Estimated Unemployment Rate (%)']))]
```

```python
### We will plot a pie chart also
plt.figure(figsize=(10,10))
plt.pie(a['Estimated Unemployment Rate (%)'],labels=a['Region'],autopct='%1.1f%%',explode=z)
plt.title('Unemployment Share per Region',color='r',loc='center')
plt.style.use('ggplot')
plt.show()
```

```python
## Now we will perform same thing for Estimated Labour Participation Rate (%)


a=pd.DataFrame(df.groupby('Region')['Estimated Labour Participation Rate (%)'].agg('sum').reset_index())

```

```python
## Now we will plot graph b/w Estimated Labour Participation Rate (%) and Region

plt.figure(figsize=(10,6))
plt.plot(a['Region'],a['Estimated Labour Participation Rate (%)'],marker='o')
plt.xlabel('Regions')
plt.ylabel('employment Rate')
plt.title('employment Rate vs Region')
plt.xticks(rotation=90)
plt.show()
```

```python
### We will plot a pie chart also
plt.figure(figsize=(10,10))
plt.pie(a['Estimated Labour Participation Rate (%)'],labels=a['Region'],autopct='%1.1f%%')
plt.title('employment Share per Region',color='r',loc='center')
plt.style.use('ggplot')
plt.show()
```

```python
### Now We will see Estimated Labour Participation Rate (%) based on Urban and Rural

a=pd.DataFrame(df.groupby(['Area'])['Estimated Labour Participation Rate (%)'].agg('mean').reset_index())
```

```python
## We will Visualize this

import seaborn as sns
plt.figure(figsize=(10,8))
plt.bar(a['Area'],height=a['Estimated Labour Participation Rate (%)'],width=0.2)
plt.xlabel('Area')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.title('Area VS Estimated Labour Participation Rate (%)')
plt.show()
```
